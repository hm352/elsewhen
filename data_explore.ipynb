```python
import numpy as np

def find_rate_area(zipcode):
    zipcode = zip_rates[zip_rates['zipcode'] == zipcode]
    rate_areas = zipcode['rate_area'].unique()
    if rate_areas.size == 1:
        print(rate_areas)
        return np.sort(zipcode['rate'].unique())[1]

print(find_rate_area(633359))
print(find_rate_area(50014))
```
